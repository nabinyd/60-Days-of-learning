## Neural Networks

**How neural network process data?**

## Inputs
- The image shows examples of various types of inputs, such as text, audio, and images.
- Before these inputs can be used by a neural network, they need to be converted into numerical form. This process is called numerical encoding.

## Numerical Encoding
- Inputs are transformed into numerical arrays (e.g., `[[116, 78, 15], [117, 43, 96], [125, 87, 23], ...]`).
- This numerical data can then be fed into the neural network.

## Neural Network Structure
- The central part of the image depicts a simplified neural network with nodes, which are called "hidden units" or "neurons."
- The network learns representations from the numerical data. These representations are patterns, features, or weights that the network uses to make predictions or decisions.

## Representation Outputs
- The neural network processes the input data through its layers and produces representation outputs (e.g., `[[0.983, 0.004, 0.013], [0.110, 0.889, 0.001], [0.023, 0.027, 0.985], ...]`).
- These outputs are numerical values that represent the network's interpretation of the input data.

## Outputs
- The representation outputs are then converted into a form that humans can understand. For example, the neural network might classify images as "Ramen" or "Spaghetti."
- In the case of text or speech inputs, the outputs could be a textual response, such as "Not a disaster" or a command like "Hey Siri, whats the weather today?"



**Anatomy of Neural Network**

*  **Input layer**: Data goes in here
*  **Hidden layers**: learns patterns in data
*  **Output layers**: outputs learned representation or prediction probabilities

***Note : "patterns" is an arbitrary term, you'll often hear "embedding" , "weights","features representation","feature vector" all referring to similar things.***

##**Types of learning**

###**Supervised Learning**  

Supervised learning involves training model on labeled data. The model learns to map inout data to the correct output based on the provided labels. It is used for tasks where the outcome or target variable is known.

**Characteristics**
*  **Labeled data**: The training dataset include input- output pairs.
*  **Goal**: Learb a napping from inputs to outputs.
*  **Applications**: Regression, classification

**Examples**
*  Regression : Predicting continuous values (eg. predicting house prices)
*  Classification : Predicting categorical values (eg. spam detection, image classification)

**Common algorithms:**
*  Linear regression
*  Logistic Regression
*  Decision Trees
*  Random forest
*  Support vector machine
*  Neural networks

###**Unsupervised learning Or self-supervised learning**

Unsupervised learning involves training a model on data without labeled responses. The model tries to find hidden patterns or intrinsic structures in the input data.

**Characteristics**
*  Unlabeled data : The training dataset includes only input data.
*  Goal : Discover underlying structure or distributoin in data.
*  Applications : Clustering, association. dimensionality reduction.

**Examples**
*  Clustering : Grouping similar data points together (eg. customer segmentation)

*  Association : Finding rules that describe large portions of the data (eg market basket analysis)

*  Dimensionality reduction : Reducing the number of features while preserving important information (eg PCA)

 **Common algorithm**
 *  K-means clustering
 *  Hierarchical clustering
 *  DBSCAN
 *  Principal component analysis (PCA)
 *  t-distributed Stochastic Neighbour Embedding (t-SNE)

###**Reinforcement Learning**

Reinforcement Learning involves training an agent to make a sequence of decisions by interacting wiht an environment. The agent learns to achieve a goal by maximizing cumulative rewards.

**Characteristics**
*  Agent-Environment Interaction : The agent interacts with the environment and learns from the consequences of its actions.
*  Goal: Learn a policy that maximizes cumulative rewards.
*  Applications : Game playing, robotics, Autonomous vechiles.

**Examples**

*  Game playing : Training agents to play games like chess, Go, or video games.
*  Robotics : Teach robots to perform tasks like walking or grasping objects.
*  Autonomous Vehicles : Training self-driving cars to navigate roads safely.


**Common Algorithms**
*  Q-learning
*  Deep Q-learning (DQN)
*  Policy Gradient Methods
*  Actor-Critic Methods
*  Proximal Policy Optimization (PPO)
*  AlphaGo


***Summary***

*  ***Supervised Learning:*** Uses labeled data to predict outcomes (e.g., regression, classification).

*  ***Unsupervised Learning:*** Uses unlabeled data to find patterns (e.g., clustering, dimensionality reduction).

*  ***Reinforcement Learning:*** Trains an agent to make decisions through rewards and penalties (e.g., game playing, robotics).

###**Transfer Learning**

Transfer learning is a machine learning technique where a pre-trained model developed for a particular task is reused as the starting point for a model on a second task. This Technique leverages the knowledge gained from the first task to improve learning efficiency and performance on the second task. Transfer learning is especially useful when the second task has limited labeled data but is releted to the first task.

**Key concept of Transfer learning**

1. Pre-trained model
   *  A model that has been already trained on large dataset, usually for a similar or related task.

   *  Commonly used pre-trained models include VGG, RestNet, Inception,BERT, and GPT. for various applications like image classification and natural language processing.

2. Feature Extraction:

   *  Using the pre-trained model to ectract meaningful features from the new dataset.

   *  The pretrained model's layers serves as a feature extractors, and only the final layer's may be re-trained to adapt to the new task.

3. Fine-tuning:
   
   *  Slightly adjusting the weights of the pre-trained model by continuing the training process on the new task's dataset.

   *  This improve performance on the new task by adapting the moldel more closely to the sprecific characteristics of the new data.


**Application of Transfer Learning**

*  Image Classification:
   *  Using a model pre-trained on ImageNet to classify different types of images with a smaller dataset.

*  Natural language processesing:
   *  Using models like BERT or GPT, which have been pre-trained on large text corpora, to perform tasks such as sentiment analysis, text classification, and queestion answering.

*  Speech recognition:
  *  Applying models pre-trained on large speech datasets to recoginze and transcribe new speech data with limited labeled examples.

###Introduction to Tensors

**Creating tensors**

PyTorch tensors are created using torch.Tensor()

**ADvantages of Tensors**
  *  **GPU acceleration** : PyTorch tensor can leverage the processing power of GPU, Significantly speeding up computations compared to CPUs.
  *  **Automatic differentiation** : Tensors are designed to work seamlessly with PyTorch's automatic differentiation (Autograd)  engine, which is essential for training neural networks.
      *  Autograd automatically calulates gradients, simplifying the process of optimizing model paramentes.
  *  **Dynamic Computation** : Tensors can be resized and reshaped during execution, allowing for more flexible model architectures.


**Tensor Gradients:**

*  A tensor gradient represent the rate of change of a scalar value (often the loss function) with respect to each element in a tensor.

*  Imagine you have a model that takes an input image and predicta a class (e.g. cat or dog). The loss function tells you how wrong your prediction is.

*  The tensor gradient tells you how much each pixel in the input image contributes to the over all error (loss)

**Autograd** (*automatic Differentiation*)

* Autograd is a powerful feature inpytorch that automatically calculaes these tensor gradients for you.

* It eliminates the neede to manually compute gradients, which can be tedius and error-prone for complex models.


***Summary***
*  Tensor gradients tell you how much each element in tensor influences the overall error (loss).
*  Autograd automates the calculations of these gradients, making deep learning training more efficient.

### Devices (`device`)

Tensors can be allocated on different devices. The most common devices are CPU and GPU (CUDA).

- `torch.device('cpu')`
- `torch.device('cuda')`
- `torch.device('cuda:0')`
- `torch.device('cuda:1')`
- ... (for multiple GPUs, use `cuda:2`, `cuda:3`, etc.)

### Gradient Calculation (`requires_grad`)

The requires_grad parameter is a boolean that indicates whether autograd should record operations on the tensor. This is useful for backpropagation during training.

- `requires_grad=True`
- `requires_grad=False`

### Pinned Memory (`pin_memory`)

The pin_memory parameter is a boolean that indicates whether the tensor should be allocated in pinned (page-locked) memory. This is useful for faster data transfer between CPU and GPU.

- `pin_memory=True`
- `pin_memory=False`


```python
float_32_tensor = torch.tensor([3,6,9],dtype=None,   # what data dype is tensor
                               device=None,  # What device is your tensor on
                               requires_grad=False,  # wether or not to track gradients with this tenosrs operation
                               pin_memory=False)
float_32_tensor
```
